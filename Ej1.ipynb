
## A) Implementar un Generador Congruencial Lineal (GCL) de módulo 2^32, multiplicador 1013904223, incremento de 1664525 y semilla igual a la parte entera del promedio de los números de padrón de los integrantes del grupo.

## El Generador Lineal Congruente tiene la forma: 

#X(n+1) = (aX(n) + c) mod m

*   m el modulo
*   a el multiplicador
*   c el incremento
*   X(0) la semilla

###En este ejercicio nos dan los valores de m=2^32, a=1013904223, c=1664525, y por ultimo X(0) que hay que calcular.

```
semilla = int((101187 + 102323 + 102291 + 102321) / 4)
Xo = semilla
print(semilla)
```

```
def semilla_gcl(semilla=102030):
  global Xo
  Xo = semilla
```

```
def generador_congruencial_lineal(iteraciones):
  m=2**32
  a=1013904223
  c=1664525
  global Xo
  
  numeros_aleatorios = []
  for i in range(0,iteraciones):
    X = (a * Xo + c) % m
    numeros_aleatorios.append(X)
    Xo = X
  return numeros_aleatorios
```

## B) Modificar el GCL implementado en el punto a) para que devuelva números al azar entre 0 y 1.

###Para que el GCL devuelva valores entre 0 y 1, lo unico que hay que modificar es que luego de calcular el X lo dividido por el módulo(m).

```
def generador_congruencial_lineal_normalizado(iteraciones):
  m=2**32
  a=1013904223
  c=1664525
  global Xo

  numeros_aleatorios = []
  for i in range(0,iteraciones):
    X = (a * Xo + c) % m
    numeros_aleatorios.append(X/m)
    Xo = X
  return numeros_aleatorios
```

## C) Realizar los gráficos que considere necesarios para mostrar las distribuciones de números al azar generados en los puntos a) y b)

### Como sabemos, las distribuciones de numeros al azar generados por un GCL tiene distribucion uniforme. Para demostarlo, vamos a graficar un histograma, debera cumplirse que todas sus barras sean aproximadamente del mismo largo.



### Primero vamos a probar el GCL sin normalizacion. 

```
numeros_al_azar = generador_congruencial_lineal(10000)
```

```
num_bins = 20
plt.figure(figsize=[30,25])
plt.hist(numeros_al_azar, num_bins, facecolor="pink", alpha=0.5)
plt.xlabel('Posible valores', fontsize=28)
plt.ylabel('Frecuencia', fontsize=28)
plt.title(r'Histograma de distribucion Uniforme', fontsize=40)
plt.xticks(fontsize=28)
plt.yticks(fontsize=28)
plt.show()
```

### Ahora vamos a probar el GCL normalizado entre los valores 0 y 1. 

```
numeros_al_azar_normalizado = generador_congruencial_lineal_normalizado(100000)
```

```
num_bins = 20
plt.figure(figsize=[30,25])
plt.hist(numeros_al_azar_normalizado, num_bins, facecolor="orange", alpha=0.5)
plt.xlabel('Posible valores', fontsize=28)
plt.ylabel('Frecuencia', fontsize=28)
plt.title(r'Histograma de distribucion Uniforme', fontsize=40)
plt.xticks(fontsize=28)
plt.yticks(fontsize=28)
plt.show()
```

###Apesar de que no son exactamente todas las barras del mismo tamaño, son todas muy parecidas, por lo tanto tiene una distribucion uniforme.
